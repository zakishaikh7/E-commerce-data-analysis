```python
values = df['productname'].value_counts().values
labels = df['productname'].value_counts().index
plt.figure(figsize=(10, 6))
explode = (0.1,) * len(labels) 
colors = ['red', 'green', 'blue', 'yellow', 'purple', 'orange'] 
plt.pie(values, labels=labels, autopct="%.2f%%", normalize=True, shadow=True, explode=explode, colors=colors)
legend = plt.legend(labels, loc='upper right')
legend.set_bbox_to_anchor((1.3, 1.3))
plt.title("productname")
plt.show()
```

```python
values = df['area'].value_counts().values
labels = df['area'].value_counts().index
plt.figure(figsize=(10, 6))
explode = (0.1,) * len(labels) 
colors = ['red', 'green', 'blue', 'yellow', 'purple', 'orange'] 
plt.pie(values, labels=labels, autopct="%.2f%%", normalize=True, shadow=True, explode=explode, colors=colors)
legend = plt.legend(labels, loc='upper right')
legend.set_bbox_to_anchor((1.3, 1.3))
plt.title("area")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X=df['productname']
y=df['productprice']
X_train,X_test,y_train,y_test = train_test_split(X,y , test_size=0.2)
```

```python
X_test_reshaped = X_test.values.reshape(-1, 1)
```
